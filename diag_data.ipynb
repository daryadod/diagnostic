```python
z_diag_id = clean_df.query("block_id.str.startswith('(Z')") \
    .groupby(['block_id', 'block_name'], as_index=False)\
    .aggregate({'id':'count'}) \
    .sort_values('id', ascending=False) \
    .rename(columns={'block_id':'z_block_id', 'block_name':'z_block_name', 'id':'count_diag'})

z_diag_id
```

```python
# groupby --> speciality -- extentdiag - кто из врачей заполняет и не заполняет заключения
ex = clean_df.query("extentdiag == 'Не указано' ") \

ex_special = ex.groupby(['speciality', 'extentdiag'], as_index=False) \
    .aggregate({'id':'count'}) \
    .sort_values('id', ascending=False) \
    .rename(columns={'extentdiag':'status_ex', 'id':'count_ex'})
ex_special
```

```python
# groupby --> популярные диагнозы (diag_id, diag_name) 

oculist = clean_df.query("speciality == 'Офтальмолог' ") \
    .groupby(['diag_id', 'diag_name'], as_index=False) \
    .aggregate({'id':'count'}) \
    .sort_values('id', ascending=False) \
    .rename(columns={'diag_id':'oculist_diag_id', 'diag_name': 'oculist_diag_name'})

persent_o= round(oculist.id *100 / oculist.id.sum(), 2)

oculist =oculist.merge(persent_o, left_index=True, right_index=True) \
.rename(columns= {'id_x': 'diag_count', 'id_y': 'diag %'})
oculist
```

```python
# groupby --> популярные диагнозы (diag_id, diag_name) 

neurologist = clean_df.query("speciality == 'Невролог'") \
    .groupby(['diag_id', 'diag_name'], as_index=False) \
    .aggregate({'id':'count'}) \
    .sort_values('id', ascending=False) \
    .rename(columns={'diag_id':'neurologist_diag_id', 'diag_name': 'neurologist_diag_name'})

persent_n= round(neurologist.id *100 / neurologist.id.sum(), 2)

neurologist.merge(persent_n, left_index=True, right_index=True) \
.rename(columns= {'id_x': 'diag_count', 'id_y': 'diag %'})
```

```python
# groupby --> популярные диагнозы (diag_id, diag_name) 

surgeon = clean_df.query("speciality == 'Хирург' ") \
    .groupby(['diag_id', 'diag_name'], as_index=False) \
    .aggregate({'id':'count'}) \
    .sort_values('id', ascending=False) \
    .rename(columns={'diag_id':'surgeon_diag_id', 'diag_name': 'surgeon_diag_name'})

persent_s= round(surgeon.id *100 / surgeon.id.sum(), 2)

surgeon.merge(persent_s, left_index=True, right_index=True) \
.rename(columns= {'id_x': 'diag_count', 'id_y': 'diag %'})
```

```python
# groupby --> популярные диагнозы (diag_id, diag_name) 

pediatric_surgeon = clean_df.query("speciality == 'Детский хирург' ") \
    .groupby(['diag_id', 'diag_name'], as_index=False) \
    .aggregate({'id':'count'}) \
    .sort_values('id', ascending=False) \
    .rename(columns={'diag_id':'pediatric_surgeon_diag_id', 'diag_name': 'pediatric_surgeon_diag_name'})

persent_ps= round(pediatric_surgeon.id *100 / pediatric_surgeon.id.sum(), 2)

pediatric_surgeon.merge(persent_ps, left_index=True, right_index=True) \
.rename(columns= {'id_x': 'diag_count', 'id_y': 'diag %'})
```

```python
# groupby --> популярные диагнозы (diag_id, diag_name) 

gastroenterologist = clean_df.query("speciality == 'Гастроэнтеролог' ") \
    .groupby(['diag_id', 'diag_name'], as_index=False) \
    .aggregate({'id':'count'}) \
    .sort_values('id', ascending=False) \
    .rename(columns={'diag_id':'gastroenterologist_diag_id', 'diag_name': 'gastroenterologist_diag_name'})

persent_g= round(gastroenterologist.id *100 / gastroenterologist.id.sum(), 2)

gastroenterologist = gastroenterologist.merge(persent_g, left_index=True, right_index=True) \
.rename(columns= {'id_x': 'diag_count', 'id_y': 'diag %'})

gastroenterologist
```

```python
# groupby --> популярные диагнозы (diag_id, diag_name) 

otolaryngologist  = clean_df.query("speciality == 'Отоларинголог' ") \
    .groupby(['diag_id', 'diag_name'], as_index=False) \
    .aggregate({'id':'count'}) \
    .sort_values('id', ascending=False) \
    .rename(columns={'diag_id':'otolaryngologist_diag_id', 'diag_name': 'otolaryngologist_diag_name'})

persent_o = round(otolaryngologist.id *100 / otolaryngologist.id.sum(), 2)


otolaryngologist.merge(persent_o, left_index=True, right_index=True) \
.rename(columns= {'id_x': 'diag_count', 'id_y': 'diag %'})
```
